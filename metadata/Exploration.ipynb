```python
import csv
import matplotlib.pyplot as plt
import pandas as pd
import xlrd
```

```python
xl = pd.ExcelFile("metadata.xlsx")
```

```python
print(xl.sheet_names)
```

```python
df = xl.parse('Sheet1')
```

```python
df.head()
```

```python
%matplotlib notebook
# distributions of readability across all current collected articles
ax = df.hist(column = ["Readability (Dejargonizer)"])
plt.title("Readability (Dejargonizer) - Rakedzon et. al 2017")
plt.axvline(95, color='r', linestyle='dashed', linewidth=1)
plt.axvline(98, color='g', linestyle='dashed', linewidth=1)

ax2 = df.hist(column= ["Readability (Flesch-Kincaid)"])
```

```python
df2 = df.groupby("Publication")["Readability (Dejargonizer)", "Readability (Flesch-Kincaid)"].mean()
df2
df2 = pd.DataFrame(df2)
ax3 = df2.plot.bar(rot = 30, ylim = (40, 100), color = ["maroon", "gold"])
```

```python
df3 = df.groupby("Field")["Readability (Dejargonizer)", "Readability (Flesch-Kincaid)"].mean()
df3
df3 = pd.DataFrame(df3)
ax4 = df3.plot.bar(rot = 60, ylim = (20, 100), colormap = "plasma")
```
